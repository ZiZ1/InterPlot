## Example of  scatter_explorer usage

Below is an esample of label_point usage. Upon click, the  closest point of a scatter plot is hightlighted and the correspondend label is placed in the position of click. If `remove_on_click=True`, repeating click near the same point removes the label.

```python
%matplotlib notebook

import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
from scatter_explorer import label_point
```

```python
sample_x = [1.0,2.0,3.0,4.0]
sample_y = [2.0,3.0,4.0,5.0]
labels = ['cat','dog','fish','mouse']
fig = plt.figure()
ax = fig.add_subplot(111)
scatter_object = ax.scatter(sample_x,sample_y,color='black')
label_point(fig, ax, scatter_object,labels,remove_on_click=True)
ax.set_xlabel("X")
ax.set_ylabel("Y")
```
